```python
cut_categories=['Fair','Good','Very Good','Premium','Ideal']
encoder=OrdinalEncoder(categories=[cut_categories])
df['cut']=encoder.fit_transform(df[['cut']])+1

 
```

```python
color_categories=['D','E','F','G','H','I','J']
encoder=OrdinalEncoder(categories=[color_categories])
df['color']=encoder.fit_transform(df[['color']])+1
```

```python
clarity_categories=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
encoder=OrdinalEncoder(categories=[clarity_categories])
df['clarity']=encoder.fit_transform(df[['clarity']])+1
```

```python
X=df.drop('price',axis=1)
```

```python
df['cut']
```

```python
#numerical pipeline
num_pipeline=Pipeline(
    steps=[
        ('imputer for handling missing values',SimpleImputer(strategy='median')),
        ('scaler for feature scaling',StandardScaler())
    ]
)

## for categorical feature
categorical_pipeline=Pipeline(
    steps=[
        ('imputer for handling the missing values' ,SimpleImputer(strategy='most_frequent')),
            
        ('scaler',StandardScaler())
        ]
)

preprocessor=ColumnTransformer([
    ('num_pipeline',num_pipeline,numericals),
    ('categorical_pipelines',categorical_pipeline,categoricals)
])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30)

```

```python

X_train
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
import numpy as np
def evaluate_model(true,predicted):
    mae= mean_absolute_error(true,predicted)
    mse= mean_squared_error(true,predicted)
    rmse=np.sqrt(mean_squared_error(true,predicted))
    r2_squaree=r2_score(true,predicted)
    return mae,mse,rmse,r2_squaree
```

```python
models={
    'LinearRegression':LinearRegression(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet(),
    'LassoR':Lasso()
}
model_list=[]
r2_list=[]
for i in range(len(list(models))):
    model=list(models.values())[i]

    model.fit(X_train,y_train)

    #predicting
    y_pred=model.predict(X_test)

   #evaluate_model=(y_test,y_pred)
    # evaluate_model
    maee ,msee,rmsee, r2_squaree= evaluate_model(y_test,y_pred)
    print(list(models.keys())[i])
    

    print("Model training performance")
    print("rmse",rmsee)
    print("mae",maee)
    print("r2_square,",r2_squaree)

    r2_list.append(r2_squaree)
```
