```python
df=pd.read_csv('diamonds.csv')



df.head()

```

```python

df.columns.str.match("Unnamed")
df=df.drop("Unnamed: 0",axis=1)
```

```python
X=df.drop("price",axis=1)    #seggregating independent features and dependent features
Y=df[['price']]
```

```python
#defining which columns should be ordinal encoded and which should be 
categoricals=X.select_dtypes(include='object').columns
numericals=X.select_dtypes(exclude='object').columns

```

```python
numericals
```

```python
#defining the custom  ranking from high to low of each ordinal feature
cut_categories=['Fair','Good','Very Good','Premium','Ideal']
clarity_categories=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
color_categories=['D','E','F','G','H','I','J']
```

```python
from sklearn.impute import SimpleImputer #for handling the misssing values 
from sklearn.preprocessing import StandardScaler# for feature scaling
from sklearn.preprocessing import OrdinalEncoder #for ordinal encoding to categorical variables.

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
#numerical pipeline
num_pipeline=Pipeline(
    steps=[
        ('imputer for handling missing values',SimpleImputer(strategy='median')),
        ('scaler for feature scaling',StandardScaler())
    ]
)

## for categorical feature
categorical_pipeline=Pipeline(
    steps=[
        ('imputer for handling the missing values' ,SimpleImputer(strategy='most_frequent')),
        ('ordinal enoding',OrdinalEncoder(categories=[cut_categories,clarity_categories,color_categories])),
        ('scaler',StandardScaler())
    ]
)

preprocessor=ColumnTransformer([
    ('num_pipeline',num_pipeline,numericals),
    ('categorical_pipelines',categorical_pipeline,categoricals)
])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=33)

```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```
